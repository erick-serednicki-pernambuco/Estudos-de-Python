```python
#Questão 1
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
notas = {'Matemática': [60, 58, 73, 51, 54, 75, 48, 72, 75, 83, 62, 52], 'Música': [80, 62, 70, 83, 62, 92, 79, 88, 54, 82, 64, 69]}
r = pd.DataFrame(notas, columns = ['Matemática', 'Música']) # cria uma tabela (notas) a partir do dicionario e nomeia as colunas
coeficitente_r = r.corr() # o metodo .corr() calcula o coeficiente correlação de r 
sn.heatmap(coeficitente_r, annot=True) #plota um grafico de calor quanto mais quente(mais claro ou proximo de 1) mais forte a relação
plt.show() #imprime o grafico

#Após feita a nalise foi obitdo o valor de 0.17 o que indica uma interação desprezivel entre as matérias, ou seja elas não teem influencia entre si. 
```
